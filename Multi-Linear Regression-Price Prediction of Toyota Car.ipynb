## Model Building

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
##Age	KM	HP	cc	Doors	Gears	Tax	WT
```

```python
model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+Tax+WT',data=toyo_data2).fit()
```

## Model Testing

```python
model.params
```

```python
## Finding t-vallues and p-values

model.tvalues,model.pvalues
```

```python
## r-squared and adj r squared

(model.rsquared , model.rsquared_adj) ##Model Accuracy is 86.17%
```

```python
## Building SLR and MLR Models for insignificant values 'CC' and 'Doors'

slr_cc=smf.ols('Price~cc',data=toyo_data2).fit()
slr_cc.tvalues,slr_cc.pvalues
```

```python
slr_dr=smf.ols('Price~Doors',data=toyo_data2).fit()
slr_dr.tvalues,slr_dr.pvalues
```

```python
## p-value is > 0.05 i.e. these variables are insignificant.

mlr_cd=smf.ols('Price~cc+Doors',data=toyo_data2).fit()
mlr_cd.tvalues,mlr_cd.pvalues
```

## Model Validation

### Collinearity

```python
rsq_age=smf.ols('Age~KM+HP+cc+Doors+Gears+Tax+WT',data=toyo_data2).fit().rsquared
vif_age=1/(1-rsq_age)

rsq_km=smf.ols('KM~Age+HP+cc+Doors+Gears+Tax+WT',data=toyo_data2).fit().rsquared
vif_km=1/(1-rsq_km)

rsq_hp=smf.ols('HP~KM+Age+cc+Doors+Gears+Tax+WT',data=toyo_data2).fit().rsquared
vif_hp=1/(1-rsq_hp)

rsq_cc=smf.ols('cc~KM+HP+Age+Doors+Gears+Tax+WT',data=toyo_data2).fit().rsquared
vif_cc=1/(1-rsq_cc)

rsq_doors=smf.ols('Doors~KM+HP+cc+Age+Gears+Tax+WT',data=toyo_data2).fit().rsquared
vif_doors=1/(1-rsq_doors)

rsq_gears=smf.ols('Gears~KM+HP+cc+Doors+Age+Tax+WT',data=toyo_data2).fit().rsquared
vif_gears=1/(1-rsq_gears)

rsq_tax=smf.ols('Tax~KM+HP+cc+Doors+Age+Gears+WT',data=toyo_data2).fit().rsquared
vif_tax=1/(1-rsq_tax)

rsq_wt=smf.ols('WT~KM+HP+cc+Doors+Age+Tax+Gears',data=toyo_data2).fit().rsquared
vif_wt=1/(1-rsq_wt)



d1={'Variables':['Age','KM','HP','CC','Doors','Gears','Tax','WT'],
   'VIF':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_tax,vif_wt]}

vif_frame=pd.DataFrame(d1)
vif_frame
```

## Model Deletion Diagnopstics and Final Model

```python
while np.max(c)>0.5 :
    model1=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+Tax+WT',data=toyo_data3).fit()
    (c,_)=model1.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    toyo_data3=toyo_data3.drop(toyo_data3.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    toyo_data3
else:
    final_model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+Tax+WT',data=toyo_data3).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
